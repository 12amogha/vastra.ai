```python
import pandas as pd
import numpy as np
import random

# Load the existing CSV file into a DataFrame
csv_file_path = 'influencer_data.csv'
df = pd.read_csv(csv_file_path)

styles = [
    'Sports and Active Wear', 'Business Formals', 'Ethnic Casuals',
    'Maternity', 'Indian and Fusion Wear', 'Western Wear', 'Traditional Wear',
    'Party Wear', 'Casual Wear', 'Formal Wear', 'Beachwear', 'Athleisure',
    'Streetwear', 'Bohemian Style', 'Vintage Clothing', 'Workout Gear',
    'Urban Fashion', 'Minimalist Fashion', 'Gothic Fashion', 'Retro Style'
]

# Dictionary defining relationships or categories (for future reference)
style_relations = {
    'Sports and Active Wear': ['Athleisure', 'Streetwear', 'Casual Wear'],
    'Business Formals': ['Formal Wear', 'Urban Fashion', 'Minimalist Fashion'],
    'Ethnic Casuals': ['Indian and Fusion Wear', 'Traditional Wear'],
    'Maternity': ['Casual Wear', 'Ethnic Casuals'],
    'Indian and Fusion Wear': ['Ethnic Casuals', 'Traditional Wear'],
    'Western Wear': ['Business Formals', 'Streetwear', 'Casual Wear'],
    'Traditional Wear': ['Indian and Fusion Wear', 'Ethnic Casuals'],
    'Party Wear': ['Casual Wear', 'Western Wear'],
    'Casual Wear': ['Sports and Active Wear', 'Streetwear', 'Western Wear'],
    'Formal Wear': ['Business Formals', 'Urban Fashion'],
    'Beachwear': ['Casual Wear', 'Streetwear'],
    'Athleisure': ['Sports and Active Wear', 'Streetwear'],
    'Streetwear': ['Sports and Active Wear', 'Casual Wear'],
    'Bohemian Style': ['Vintage Clothing', 'Retro Style'],
    'Vintage Clothing': ['Bohemian Style', 'Retro Style'],
    'Workout Gear': ['Sports and Active Wear', 'Athleisure'],
    'Urban Fashion': ['Business Formals', 'Formal Wear'],
    'Minimalist Fashion': ['Business Formals', 'Formal Wear'],
    'Gothic Fashion': ['Streetwear', 'Casual Wear'],
    'Retro Style': ['Vintage Clothing', 'Bohemian Style']
}

# Function to generate n based on left-skewed normal distribution
def generate_n():
    mean_n = 2  # Mean of the distribution
    std_dev_n = 1  # Standard deviation of the distribution
    n = np.random.normal(mean_n, std_dev_n, 1)[0]
    n = max(1, int(np.round(n)))  # Ensure n is at least 1 and round to nearest integer
    return n

# Function to select styles based on initial random style and left-skewed n
def select_styles_random_initial(n):
    initial_style = random.choice(styles)
    selected_styles = [initial_style]
    current_style = initial_style

    for _ in range(1, n):
        if current_style in style_relations:
            related_styles = style_relations[current_style]
        else:
            related_styles = styles  # Default to all styles if no specific relations

        # Filter out already selected styles
        available_styles = [style for style in related_styles if style not in selected_styles]

        # If no available styles, pick randomly from related styles
        if not available_styles:
            available_styles = related_styles

        # Choose a style from available related styles
        next_style = random.choice(available_styles)
        selected_styles.append(next_style)
        current_style = next_style

    return selected_styles


# Add style column to DataFrame
df['style'] = df.apply(lambda row: select_styles_random_initial(generate_n()), axis=1)

# Save the updated DataFrame back to the same CSV file
df.to_csv(csv_file_path, index=False)

# Display the final DataFrame
print(df)
```

```python
import pandas as pd
import random

# Load the existing CSV file into a DataFrame
csv_file_path = 'influencer_data.csv'
df = pd.read_csv(csv_file_path)

# Define the brands and styles mapping
brands = ['Allen Solly', 'Levi\'s', 'Adidas', 'Gucci', 'H&M', 'Under Armour', 'Puma', 'Nike', 'Zara', 'Reebok']
style_to_brands = {
    'Sports and Active Wear': ['Adidas', 'Under Armour', 'Puma', 'Nike', 'Reebok'],
    'Business Formals': ['Allen Solly', 'Levi\'s', 'Gucci', 'Zara'],
    'Ethnic Casuals': ['Allen Solly', 'Zara'],
    'Maternity': ['H&M'],
    'Indian and Fusion Wear': ['Allen Solly', 'Zara'],
    'Western Wear': ['Levi\'s', 'Gucci', 'H&M', 'Zara'],
    'Traditional Wear': ['Allen Solly'],
    'Party Wear': ['Gucci', 'Zara'],
    'Casual Wear': ['Levi\'s', 'H&M', 'Zara'],
    'Formal Wear': ['Allen Solly', 'Gucci', 'Zara'],
    'Beachwear': ['H&M'],
    'Athleisure': ['Adidas', 'Under Armour', 'Puma', 'Nike', 'Reebok'],
    'Streetwear': ['Levi\'s', 'H&M', 'Zara'],
    'Bohemian Style': ['Zara'],
    'Vintage Clothing': ['Zara'],
    'Workout Gear': ['Adidas', 'Under Armour', 'Puma', 'Nike', 'Reebok'],
    'Urban Fashion': ['Levi\'s', 'Gucci', 'Zara'],
    'Minimalist Fashion': ['Levi\'s', 'Gucci', 'Zara'],
    'Gothic Fashion': ['Zara'],
    'Retro Style': ['Zara']
}

def get_related_brands(styles_list):
    related_brands = set()
    for style in styles_list:
        style = style.strip().strip('"').strip('[').strip(']').strip("'")  # Clean unwanted characters
        if style in style_to_brands:
            related_brands.update(style_to_brands[style])
        else:
            print(f"Style '{style}' not found in style_to_brands dictionary.")

    related_brands = list(related_brands)
    if not related_brands:
        return []

    n = random.randint(1, len(related_brands)) if related_brands else 0
    return random.sample(related_brands, n) if n > 0 else []

# Clean and process the 'style' column
def parse_styles(style):
    if pd.isna(style):
        return []
    if isinstance(style, str):
        # Remove unwanted characters and split by commas
        cleaned_style = style.strip().strip('[]').replace('"', '').replace("'", '')
        return [s.strip() for s in cleaned_style.split(',') if s.strip()]
    return []

df['style'] = df['style'].apply(parse_styles)

# Apply the function to DataFrame
df['past_collab'] = df['style'].apply(lambda x: get_related_brands(x if isinstance(x, list) else [x]))

# Save the updated DataFrame back to the same CSV file
df.to_csv(csv_file_path, index=False)

# Display the final DataFrame
print(df)
```

```python
csv_file_path = 'influencer_data.csv'
df = pd.read_csv(csv_file_path)

# Define weights for each factor
weights = {
    'gender': 0.6,
    'follower_locations': 0.25,
    'style': 0.05,
    'age_range': 0.1,
}

def calculate_trend_alignment(row, location, style_criteria, age_range_criteria, gender_criteria):
    # Initialize score
    score = 0

    # Gender alignment check
    if row['gender'] == gender_criteria or gender_criteria == 'Unisex':
        score += weights['gender']
    else:
        score -= weights['gender']

    # Follower locations alignment check
    if location in row['follower_locations']:
        score += weights['follower_locations']
    else:
        score -= weights['follower_locations']

    # Style alignment check
    common_styles = set(row['style']).intersection(style_criteria)
    style_alignment = len(common_styles) / len(style_criteria) if style_criteria else 0
    score += weights['style'] * style_alignment

    # Age range alignment check
    if row['age_range'] == age_range_criteria:
        score += weights['age_range']
    else:
        # Calculate age_range weight based on difference
        min_age, max_age = map(int, age_range_criteria.split('-'))
        influencer_min_age, influencer_max_age = map(int, row['age_range'].split('-'))

        if influencer_min_age <= min_age <= influencer_max_age or \
           influencer_min_age <= max_age <= influencer_max_age:
            age_weight = 1.0
        else:
            age_weight = 0.5

        score += weights['age_range'] * age_weight

    return score

# Example criteria
location_criteria = 'Jaipur'
style_criteria = ['Streetwear', 'Casual Wear']
age_range_criteria = '25-34'
gender_criteria = 'Male'

# Calculate trend alignment score for each influencer
df['trend_alignment_score'] = df.apply(lambda row: calculate_trend_alignment(row, location_criteria, style_criteria, age_range_criteria, gender_criteria), axis=1)

# Normalize trend_alignment_score to range [-1, 1]
max_score = df['trend_alignment_score'].max()
min_score = df['trend_alignment_score'].min()
df['trend_alignment_score'] = (2 * (df['trend_alignment_score'] - min_score) / (max_score - min_score)) - 1

# Sort by trend_alignment_score in descending order
df = df.sort_values(by='trend_alignment_score', ascending=False)


# Display the final DataFrame with trend alignment scores
print(df[['influencer_name', 'gender', 'follower_locations', 'style', 'age_range', 'trend_alignment_score']])

# Save the updated DataFrame back to the same CSV file
df.to_csv(csv_file_path, index=False)
```

```python
csv_file_path = 'influencer_data.csv'
df = pd.read_csv(csv_file_path)

def calculate_brand_fit_score(row):
    # Define weights for each brand value
    weights = {
        'customer_obsession': 0.25,
        'operational_excellence': 0.15,
        'innovation': 0.20,
        'inclusivity': 0.20,
        'sustainability': 0.20
    }

    # Normalize followers and engagement_metrics
    max_followers = df['followers'].max()
    max_engagement_metrics = df['engagement_metrics'].max()
    
    customer_obsession_score = (row['followers'] / max_followers) * 10 + (row['engagement_metrics'] / max_engagement_metrics) * 10

    # Operational excellence score based on public perception and past collaborations
    public_perception_score = 10 if row['public_perception'] == 'mostly positive' else 5 if row['public_perception'] == 'neutral' else 0
    operational_excellence_score = public_perception_score + len(eval(row['past_collab']))

    # Innovation score based on style
    innovation_score = len(eval(row['style']))

    # Inclusivity score based on follower locations
    inclusivity_score = len(eval(row['follower_locations']))

    # Sustainability score based on eco-friendly past collaborations
    eco_friendly_brands = ['Nike', 'Puma', 'Levi\'s', 'H&M', 'Zara']
    sustainability_score = sum(1 for brand in eval(row['past_collab']) if brand in eco_friendly_brands)

    # Calculate weighted score
    scores = {
        'customer_obsession': customer_obsession_score,
        'operational_excellence': operational_excellence_score,
        'innovation': innovation_score,
        'inclusivity': inclusivity_score,
        'sustainability': sustainability_score
    }
    
    brand_fit_score = sum(weights[key] * scores[key] for key in weights)
    return brand_fit_score

# Apply the scoring function to each row
df['brand_fit_score'] = df.apply(calculate_brand_fit_score, axis=1)
df[['influencer_name', 'brand_fit_score']].sort_values(by='brand_fit_score', ascending=False)

# Save the updated DataFrame back to the same CSV file
df.to_csv(csv_file_path, index=False)
print(df)
```
